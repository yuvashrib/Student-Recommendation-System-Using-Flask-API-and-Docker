Fetch JSON data from multiple API URLs and saves each response to a local file (e.g., data_1.json) ensuring error handling during the download and formats the data for easy use later.

```python
import requests
import json

# List of API URLs
api_urls = [
    "https://api.jsonserve.com/rJvd7g",
    "https://www.jsonkeeper.com/b/LLQT",
    "https://api.jsonserve.com/XgAgFJ"
]

# Function to download data from API and save it to a JSON file
def download_and_save_json(url, filename):
    try:

        response = requests.get(url)
        response.raise_for_status()


        data = response.json()


        with open(filename, 'w') as f:
            json.dump(data, f, indent=4)

        print(f"Data from {url} saved to {filename}")

    except requests.exceptions.RequestException as e:
        print(f"Error downloading data from {url}: {e}")


for i, url in enumerate(api_urls, start=1):
    filename = f"data_{i}.json"
    download_and_save_json(url, filename)
```

After downloading, these JSON files can be converted into Excel files for better organization and readability. Once the Excel files are created, they can be loaded back into a Python IDE for further data processing and analysis.

```python
# Load the Historical data csv file
historical_data_path = "Historical_Data.csv"
historical_data_df = pd.read_csv(historical_data_path,encoding="latin1")
print("Historical Data DataFrame:")
historical_data_df.head()
```

```python
# Drop any duplicates
historical_data_df.drop_duplicates(inplace=True)
```

```python
# Standardize the column names for easy access
historical_data_df.columns = historical_data_df.columns.str.lower().str.replace(' ', '_')
historical_data_df
```

```python
# Segregate Categorical Data, Numerical data and date columns
categorical_cols = historical_data_df.select_dtypes(include=['object']).columns
numeric_cols = historical_data_df.select_dtypes(include=['int64', 'float64']).columns
date_cols = historical_data_df.select_dtypes(include=['datetime64']).columns
```

```python
# Fill the null values
historical_data_df[categorical_cols] = historical_data_df[categorical_cols].fillna('Unknown')
historical_data_df[numeric_cols] = historical_data_df[numeric_cols].fillna(historical_data_df[numeric_cols].median())
for col in date_cols:
    historical_data_df[col] = historical_data_df[col].fillna(historical_data_df[col].min())
```

```python
print("Missing values after cleaning:")
historical_data_df.isnull().sum()
```

```python
historical_data_df.columns = historical_data_df.columns.str.replace(r'^.*?\.', '', regex=True)
historical_data_df
```

```python
historical_data_df.columns
```

Group quiz data by quiz_id to calculate aggregated totals for total_questions, correct_answers, and incorrect_answers. Then computes the number of unanswered questions for each quiz by subtracting correct and incorrect answers from the total questions. The result is a summarized dataset showing quiz performance metrics for each quiz.

```python
grouped_data = historical_data_df.groupby('quiz_id').agg({
    'total_questions': 'sum',
    'correct_answers': 'sum',
    'incorrect_answers': 'sum'
}).reset_index()

grouped_data['unanswered_questions'] = (
    grouped_data['total_questions']
    - grouped_data['correct_answers']
    - grouped_data['incorrect_answers']
)

grouped_data
```

```python
grouped_data.set_index('quiz_id')[['correct_answers', 'incorrect_answers', 'unanswered_questions']].plot(
    kind='bar',
    stacked=True,
    figsize=(10, 6)
)

# Customize the chart
plt.title('Quiz Performance by Quiz ID')
plt.xlabel('Quiz ID')
plt.ylabel('Number of Questions')
plt.xticks(rotation=0)
plt.legend(title='Question Type')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
historical_data_df['quiz.topic'] = historical_data_df['quiz.topic'].str.lower().str.strip()
historical_data_df
```

```python
grouped_data = historical_data_df.groupby('quiz.topic').agg({
    'total_questions': 'sum',
    'correct_answers': 'sum',
    'incorrect_answers': 'sum'
}).reset_index()

grouped_data['unanswered_questions'] = (
    grouped_data['total_questions']
    - grouped_data['correct_answers']
    - grouped_data['incorrect_answers']
)

grouped_data
```

```python
grouped_data.set_index('quiz.topic')[['correct_answers', 'incorrect_answers', 'unanswered_questions']].plot(
    kind='bar',
    stacked=True,
    figsize=(10, 6)
)

# Customize the chart
plt.title('Quiz Performance by Quiz Topic')
plt.xlabel('Quiz Topic')
plt.ylabel('Number of Questions')
plt.xticks(rotation=90)
plt.legend(title='Question Type')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Convert accuracy column to just numbers by removing the '%' symbol
historical_data_df['accuracy'] = historical_data_df['accuracy'].str.rstrip('%').astype(float)
historical_data_df
```

```python
historical_data_df['started_at'] = pd.to_datetime(historical_data_df['started_at'])
historical_data_df['ended_at'] = pd.to_datetime(historical_data_df['ended_at'])
historical_data_df
```

```python
# calculate the total time taken
historical_data_df['time_taken'] = (historical_data_df['ended_at'] - historical_data_df['started_at']).dt.total_seconds() / 60
historical_data_df
```

```python

# Group by 'quiz.topic' and calculate average accuracy and speed
grouped_data = historical_data_df.groupby('quiz.topic').agg({
    'accuracy': 'mean',
    'speed': 'mean',
    'time_taken': 'mean'
}).reset_index()

# Rename columns for clarity
grouped_data.rename(columns={'accuracy': 'average_accuracy(%)', 'speed': 'average_speed','time_taken': 'average_time_taken (mins)'}, inplace=True)

# Display the result
grouped_data
```
